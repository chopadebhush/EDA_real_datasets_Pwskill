OR Method We can deal with numeric and categorical columns

```python
catcol =[]
numcol =[]
for i in df.columns :
        if df[i].dtypes =="O":
            catcol.append(i)
        else:
            numcol.append(i)
print(catcol)
print("==================================================================================================")
print(numcol)
```
