`How many application type people used?`

```python
plt.figure(figsize=(12,7))
df.application_type.value_counts().plot(kind="bar",color="red")
plt.xlabel("Aplication Type")
plt.ylabel("COunt")
plt.title("Application type vs count")
plt.show()
```

`Observation & Insight`

1. Certied and online Application type used more

`Which type of agent are avilable?`

```python
df_agentState =df.agent_state.value_counts()
```

```python
## Top 10 Agent state
df_agent =df_agentState.nlargest(n=10)
df_agent
```

```python
plt.figure(figsize=(12,12))
plt.pie(x=df_agent,labels=df_agent.index,autopct="%1.1f%%")
plt.title("Agents state %")
plt.show()
```

`Observation and insights`

1. certified agent has 68.5 %
2. CA Agent has 9.3%
