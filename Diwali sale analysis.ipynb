## `Import all necessary liberies`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Read the data from Dataset & Display Top5 Record
# Here we used Encding for `latin1`,`unicode_escape` because dataset was giving error of encoding  
df =pd.read_csv("Diwali Sales Data.csv",encoding="latin1")
df.head()
```

```python
# Check Dataset shape 11251 rows & 15 columns  
df.shape
```

```python
# In these cell we have to check last 5 record of dataset
df.tail()
```

```python
# Check the information about datasets
df.info()
```

```python
# Check the null values from whole dataset
# `Status`,`unnamed1 `have null values.
df.isnull().sum()
```

```python
# Drop those columns which is having null values
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
# Check Info of the dataset
df.info()
```

```python
# Check the null Values 
# In `Amount` Column 12 null values are spread.
df.isnull().sum()
```

```python
# We have drop null values from amount column
df.dropna(inplace=True)
```

```python
# Check again null value now these values was droped
df.isnull().sum()
```

```python
# Change the datatype of Amount Column
df["Amount"]=df["Amount"].astype("int")
```

## `Exploratory data analysis`

## `Gender feature Analysis :`

```python
df.columns
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x='Gender',data=df,hue="Gender")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sales =df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending= False)
```

```python
Sales
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x='Gender',y='Amount',data=Sales)
```

# **`Observation`**
As per the analysis of base on **Gender,Amount** feature we have shown female did more shopping as well
as theire expenditure is more than male.

## `AGE feature Analysis :`

```python
df.columns
```

```python
plt.figure(figsize=(7,4))
sns.set_theme(style="whitegrid")
ax= sns.countplot(x="Age Group",hue='Gender',data=df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sale_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x='Age Group',
    y='Amount',
    hue ='Age Group',
    data=Sale_age)
```

## **`Observation`**
As per the analysis on feature **age & amount** age group 25-36 Female and male did more shopping.
and Age group 25-36 spend more money on shopping.

## `State feature analysis :`

```python
df.columns
```

```python
# How many sales did in which state
plt.figure(figsize=(10,5))
sns.countplot(x=df['State'].head(20),data=df)
plt.xlabel("State",fontsize =15)
plt.ylabel("Sales Count",fontsize =15)
```

```python
#Total numbers of order from states
sale_state =df.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
```

```python
plt.figure(figsize=(15,5))
sns.set_theme(style="whitegrid")
sns.barplot(x='State',
    y='Orders',
    data=sale_state)
```

```python
# Total sale State/Amount from top 10 State 
sale_amount_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(15,4)})
sns.barplot(x='State',
    y='Amount',
    data=sale_amount_state)
```

## `Observation :`
1- As per the analysis base on **state** In Andhrapradesh & utterpradesh peoples did more shopping as compared to other.

2- The more numbers of order from utterpradesh & Maharshtra state.

3- The utterpradesh & Maharashtra did lots of expenditure on sale as per shown in figure.

## `Marital_Status feature analysis :`

```python
df.columns
```

```python
sns.set_theme(context='notebook',
    style='whitegrid',
    palette='deep',
    font='sans-serif',
    font_scale=1,
    color_codes=True)

plt.figure(figsize=(5,3))
ax =sns.countplot(x='Marital_Status',data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

## `Product category feature analysis :`

```python
df.columns
```

```python
# Most selling product
plt.figure(figsize=(25,5))

ax=sns.countplot(x=df['Product_Category'],
              data=df,palette='Set1')
plt.xlabel('Product_Category',fontsize=15)
plt.ylabel('count',fontsize=15)

for bar in ax.containers:
    ax.bar_label(bar)
```

## **`Observation`**
1 - As per countplot we have seen clothes,food,electronic gadgets had got good sales.<br>
2 - Second bar plot shown people have spend more meoney on food then clothes & electronic gadgets   based on amount feature analyssis.
