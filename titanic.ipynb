`Seprate object and numerical columns`

```python
cat_col =[i for i in df.columns if df[i].dtypes == "O"]
num_col =[i for i in df.columns if df[i].dtypes != "O"]
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="Sex",data=df)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Sex vs Count")
plt.show()
```

```python
cat_col
```

```python
fg,ax =plt.subplots(nrows=4 ,ncols=1,figsize=(12,20))
plt.subplots_adjust(hspace=0.5)

for i,col in enumerate(cat_col) :
    sns.countplot(x =col,data=df,ax=ax[i])
    
    ax[i].set_title(col,fontsize=20)
plt.show()
```

```python
num_col
```

```python
fig ,ax =plt.subplots(nrows=len(num_col),ncols=1,figsize=(12,20))
plt.subplots_adjust(hspace=.8)

for i,col in enumerate(num_col):
    sns.histplot(x=col ,data=df,kde=True,hue="Sex",ax=ax[i])
    
    ax[i].set_title(col,fontsize=20)
plt.show()
```

`How many servived male and Female `

```python
df_servived =df.Survived.value_counts()
df_servived
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="Survived",data=df,hue="Sex",)
plt.xlabel("Servived")
plt.ylabel("counts")
plt.title("Servived Male and Female")
plt.show()
```

`Insight and Observation`

1.266 male servived and 152 female servived

```python
plt.figure(figsize=(12,8))
df_servived.plot.pie(labels=df_servived.index,autopct ="%1.1f%%")
plt.title("Servived folks %")
plt.legend()
plt.show()
```
