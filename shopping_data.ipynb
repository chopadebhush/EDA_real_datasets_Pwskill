`Visualization part`

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=df["CustomerID"],y=df["Annual_income"],markers="o")
plt.title("Cust ID vs annual Income")
plt.show()
```

`which 10 customer has highest Income?`

```python
a =df.set_index("Genre")
a
```

```python
df_Annual_income= a["Annual_income"].nlargest(n=10).reset_index()
df_Annual_income
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=df["Genre"],y=df["Annual_income"])
plt.xlabel("Genre")
plt.ylabel("Annual_income")
plt.title("Genre vs Annual Income")
plt.show()
```

`Observation And Insight`

1. male has highest annual income as compare to female

`Distrubution of Age`

```python
plt.figure(figsize=(12,8))
sns.displot(x=df["Age"],kind="hist",palette="viridis")
plt.title("Age vs freq")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.histplot(x=df["Age"],kde=True)
plt.title("Age vs freq")
plt.show()
```

`Observation & Insight`

1. 20-40 age people has more count.

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x=df["Age"],y=df["Spending_Score"],markers="o")
plt.xlabel("Age")
plt.ylabel("Spending_Score")
plt.title("Age wise spending Score")
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(x=df["Age"],y=df["Spending_Score"],markers="o",)
plt.xlabel("Age")
plt.ylabel("Spending_Score")
plt.title("Age wise spending Score")
plt.show()
```

`Observation and Insight`

1. 20 to 40 age customers has 40 -75 spending score

```python
plt.figure(figsize=(12,8))
sns.barplot(x=df["Genre"],y=df["Spending_Score"])
plt.xlabel("Genre")
plt.ylabel("Spending_Score")
plt.title("Genre vs Spending_Score")
plt.show()
```

`Observation and Insight`

```python
col_df =["Genre"]

for i in col_df :
    plt.figure(figsize=(12,7))
    plt.title(i)
    sns.countplot(x=df[i],data=df,palette="Accent_r")

plt.tight_layout()   
plt.show()
```

```python
cols_num = ['CustomerID', 'Age', 'Annual_income', 'Spending_Score']

fig,ax =plt.subplots(nrows= 4 ,ncols=1,figsize=(10,20))
fig.subplots_adjust(hspace=0.5)

for i,col in enumerate(cols_num) :
    sns.boxplot(x=col ,data=df ,ax =ax[i])
    
    ax[i].set_title(col)
```
