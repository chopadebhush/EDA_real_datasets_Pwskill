`Find out unique values`

```python
def UniqueValues(col):
    
    for i in col :
        print(f"{i}==> {df[i].dtype}::==> unique values are::==>{df[i].unique()}")
        print("========================================================================================")
        
UniqueValues(df.columns)
```

`Seprate the num and cat col`

```python
cat_col =[i for i in df.columns if df[i].dtypes == "O"]
num_col =[i for i in df.columns if df[i].dtypes != "O"]
```

```python
plt.figure(figsize=(12,12))
plt.pie(x=df_add ,labels=df_add.index,autopct="%1.1f%%")
plt.title("Rural and urban student %")
plt.show()
```

`Mother and Father eduction features`

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.barplot(x=df["sex"],y=df["Medu"],hue=df.school)
plt.title("Sex vs Mother education")

plt.subplot(1,2,2)
sns.barplot(x=df["sex"],y=df["Fedu"],hue=df.school)
plt.title("Sex vs Father education")

plt.show()
```

`Observation and Insights`

1. Mother education is matter for child.
2. father education is matter for child.

`Categorical features insights`

```python
plt.figure(figsize=(20,20))

plotnumber = 1 
for i in cat_col :
    if plotnumber <= 17:
        plt.subplot(5,4,plotnumber)
        sns.countplot(df[i])
        plt.xlabel(i,fontsize=15)
        plt.title("categorical feature")
    plotnumber+=1
    
    
plt.tight_layout()  
plt.show()
```

`Numerical features insights`

```python
plt.figure(figsize=(20,20))

plotnumber = 1 
for i in num_col :
    if plotnumber <= 16:
        plt.subplot(4,4,plotnumber)
        sns.histplot(df[i])
        plt.xlabel(i,fontsize=15)
        plt.title("Numerical feature")
    plotnumber+=1
    
    
plt.tight_layout()  
plt.show()
```

`Correlation of features`

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True,vmin=-1,vmax=1,cmap="Accent")
plt.title("correlation of features",fontsize=25)
plt.show()
```
