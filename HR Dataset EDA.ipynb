#### `Find the Top 10 salary with respect to employee Name`

```python
a =df.set_index("Employee_Name")
```

```python
Top10_Highest =a["Salary"].sort_values(ascending=False).head(10)
Top10_Highest
```

```python
## We can find also Like 
a["Salary"].nlargest(n=10)
```

#### `Find Lowest top10 Salary With respet to Employee Name`

```python
Top10_Lowest =a["Salary"].sort_values(ascending=False).tail(10)
Top10_Lowest
```

```python
a["Salary"].nsmallest(n=10)
```

#### `Top 10 Highest Salry EMployee Name`

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x=Top10_Highest.index,y=Top10_Highest)
plt.xticks(rotation=90)
plt.xlabel("Employee Name")
plt.ylabel("Salary")
plt.title("Highest salary of emplyee Name")
plt.show()
```

#### `Top 10 Highest Salry EMployee Name`

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x=Top10_Lowest.index,y=Top10_Lowest)
plt.xticks(rotation=90)
plt.xlabel("Employee Name")
plt.ylabel("Salary")
plt.title("Lowest salary of emplyee Name")
plt.show()
```

#### `Distrubution of salary`

```python
plt.figure(figsize=(12,5))
sns.histplot(x=df["Salary"],kde=True)
plt.xlabel("Salary")
plt.ylabel("count")
plt.show()
```

#### `Insight and Observation`

2. Above 1L has got salary less number of folks

#### `Relationship between GenderID vs Salary`

#### `How many Position in the companyand its percent`

```python
df_Pos =df["Position"].value_counts()
```

```python
plt.figure(figsize=(25,20))
df_Pos.plot.pie(x=df["Position"],labels=df_Pos.index,autopct="%1.1f%%")
plt.title("Postion in company")
plt.legend()
plt.show()
```

#### `Observation And Insights`

1. Product Technician I & Product Technician II have more postion in organzation

```python
plt.figure(figsize=(12,7))
sns.countplot(data=df,x=df["Position"])
plt.xlabel("Postion",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```
